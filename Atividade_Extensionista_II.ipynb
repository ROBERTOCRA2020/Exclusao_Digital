```python
# Importação de bibliotacas
import pandas as pd
import numpy as np
from google.colab import files
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Importação do arquivo para tratamento e análise
dados_originais = pd.read_csv('/content/drive/MyDrive/Atividade Extensionista II - Uninter/Dados/tic_domicilios_2025_individuos_base_de_microdados_v1.0.csv', sep=';', encoding= 'utf-8')
```

```python
#visualização dos dados
dados_originais
```

```python
#Seleção das colunas para análise
dados1=dados_originais[['ID_DOMICILIO','CLASSE_2015','AREA','COD_UF','SEXO','IDADE','GRAU_INSTRUCAO_1','GRAU_INST_1','ESTUD','APOSENT','RACA','C1','C2A','B1','B2']]
```

```python
# Aplicação de filtro para selecionar somente os dados referentes a localidade do Distrito Federal
dados2 = dados_originais[dados_originais['COD_UF']==53][['ID_DOMICILIO','CLASSE_2015','AREA','COD_UF','SEXO','IDADE','GRAU_INSTRUCAO_1','GRAU_INST_1','ESTUD','APOSENT','RACA','C1','C2A','B1','B2']]
```

```python
# Alteração do nome de alugmas colunas
dados2 = dados2.rename(columns={'C1': 'USOU_INTERNET', 'C2A': 'MOTIVO', 'B1': 'USOU_COMPUTADOR', 'B2': 'QDO_USOU'})
```

```python
#Mapeamento  dos dados para subsituição
map_classe2015 = {1: 'A', 2: 'B', 3: 'C', 4: 'DE'}
map_area = {1: 'Urbana', 2: 'Rural'}
map_sexo = {1: 'Masculino', 2: 'Feminino'}
map_grinst = {1: 'Analfabeto', 2: 'Ensino Fundamental', 3: 'Ensino Médio', 4: 'Ensino Superior'}
map_estud = {1: 'Sim', 0: 'Não'}
map_aposent = {1: 'Sim', 0: 'Não'}
map_raca = {1: 'Branca', 2: 'Preta', 3: 'Amarela', 4: 'Parda', 5: 'Indígena', 98: 'Não respondeu'}
map_usouinternet = {0: 'Não', 1: 'Sim', 97: 'Não sabe', 98: 'Não respondeu'}
map_motivo = {4: 'Por não ter onde acessar', 5: 'Por ser muito caro', 6: "Por ter preocupações com segurança ou privacidade",
  7: "Por evitar o contato com conteúdo perigoso", 8: "Por outro motivo", 9: "Por falta de interesse ou necessidade",
  10: "Por não saber usar", 97: "Não sabe", 98: "Não respondeu", 99: "Não se aplica"}
map_usoucomput = {0: 'Não', 1: 'Sim', 97: 'Não sabe', 98: 'Não respondeu', 99: 'Não se aplica'}
map_qdo_usou = {1: 'Há menos de três meses', 2: 'Entre três e doze meses', 3: 'Mais de doze meses atrás', 99: "Não se aplica"}
#Aplica a substituição dos dados na coluna
dados2['CLASSE_2015'] = dados2['CLASSE_2015'].replace(map_classe2015).astype(str)
dados2['AREA'] = dados2['AREA'].replace(map_area).astype(str)
dados2['COD_UF'] = dados2['COD_UF'].replace({53: 'DF'})
dados2['SEXO'] = dados2['SEXO'].replace(map_sexo).astype(str)
dados2['GRAU_INST_1'] = dados2['GRAU_INST_1'].replace(map_grinst).astype(str)
dados2['ESTUD'] = dados2['ESTUD'].replace(map_estud).astype(str)
dados2['APOSENT'] = dados2['APOSENT'].replace(map_aposent).astype(str)
dados2['RACA'] = dados2['RACA'].replace(map_raca).astype(str)
dados2['USOU_INTERNET'] = dados2['USOU_INTERNET'].replace(map_usouinternet).astype(str)
dados2['MOTIVO'] = dados2['MOTIVO'].replace(map_motivo).astype(str)
dados2['USOU_COMPUTADOR'] = dados2['USOU_COMPUTADOR'].replace(map_usoucomput).astype(str)
dados2['QDO_USOU'] = dados2['QDO_USOU'].replace(map_qdo_usou).astype(str)
```

```python
# Salvando os dados no drive
dados2.to_csv('exclusao_digital.csv', index=False, encoding='utf-8-sig')
```
